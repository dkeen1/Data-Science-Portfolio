## PCA Lecture Practice

```python
#import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#import cancer dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer.target)
```

```python
cancer_df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
cancer_df.head()
```

### Need to scale data so that everything is on the same scale

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(cancer_df)
```

```python
scaled_data = scaler.transform(cancer_df)
```

### Actually doing the Principal Component Analysis

```python
#PCA
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
#this shows you the dimensionality of the original dataset
```

```python
x_pca.shape
#this shows you the dimensionality of the PCA dataset (which is just 2 because we only kept 2 components)
```

```python
plt.figure()
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer['target'], alpha=.5)
plt.xlabel('First Principal Component')
plt.ylabel('Sceond Principal Component')
plt.legend([0,1])
plt.show()
```

```python
pca.components_
```

```python
cancer_df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
cancer_df_comp.head()
#rows are the principal components themselves
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(cancer_df_comp, cmap='plasma')
```

Here is where your domain expertise as a data scientist comes into play, because you need to name your principal components now, so that they have more utility in your business. 

I would probably call Principal Component 1 (PC 1) something like "General Size/Shape", because we have things like mean radius, perimeter, and worst radius/perimeter showing up as negatively correlated with PC 1. 

Principal Component 0 (PC 0), I would probably call something like "Irregularity of Shape", because it seems that mean concavity and mean concave points are both positively correlated with it, and (according to my brief research), malignant tumors are typically more irregular in their shape (so higher concavity would indicate a greater likelihood something is malignant).
